## 럭키백의 확률  
---

김팀장은 럭키백에 들어갈 수 있는 생선은 7개라고 알려줌  
크기,무게 등이 주어졌을 때 7개 생선에 대한 확률을 출력해야 함  
길이,높이,두께 외에도 대각선 길이도 사용 가능

사이킷런의 k-최근점 이웃 분류기:클래스의 확률을 계산하여 제공 

### 데이터 준비하기

```python
import pandas as pd

fish = pd.read_csv('https://bit.ly/fish_csv_data')
fish.head()
```

```python
print(pd.unique(fish['Species']))
```

```python
fish_input = fish[['Weight','Length','Diagonal','Height','Width']].to_numpy()#넘파이 배열
```

```python
print(fish_input[:5])
```

```python
fish_target = fish['Species'].to_numpy()
```

```python
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(
    fish_input, fish_target, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
ss.fit(train_input)
train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)
```

### k-최근접 이웃 분류기의 확률 예측  
---

클래스 객체를 만들고 훈련 세트로 모델을 훈련하고 훈련이랑 테스트 세트의 점수 확인

```python
from sklearn.neighbors import KNeighborsClassifier

kn = KNeighborsClassifier(n_neighbors=3)
kn.fit(train_scaled, train_target)

print(kn.score(train_scaled, train_target))
print(kn.score(test_scaled, test_target))
```

타깃 데이터가 2개 이상으면 다중분류  
이진 분류이면 양성은 1, 음성은 0  
다중 분류도 숫자로 할 수 있지만, 사이킷런에서는 문자열로 된 타깃값을 사용 가능

타깃 순서는 classes_에 저장

```python
print(kn.classes_)
```

```python
print(kn.predict(test_scaled[:5]))
```

```python
import numpy as np

proba = kn.predict_proba(test_scaled[:5])
print(np.round(proba, decimals=4))
```

5개의 샘플에 대해서 5개의 클래스들의 확률

```python
distances, indexes = kn.kneighbors(test_scaled[3:4])
print(train_target[indexes])
```

## 로지스틱 회귀  
---

이름은 회귀이지만 분류 모델이다  
$z=a \times(weight)+b\times(length)+c\times(diagonal)+d\times(height)+e\times(width)+f$  
z는 범위가 (-$\infty$,$\infty$)인데, 확률이 되려면 0~1이 되어야 한다  
그래서 시그모이드 함수를 사용하여 음수일 때 0이 가까워지고 양수일 때 1에 가까워지게 함 

```python
import numpy as np
import matplotlib.pyplot as plt

z = np.arange(-5, 5, 0.1)
phi = 1 / (1 + np.exp(-z))

plt.plot(z, phi)
plt.xlabel('z')
plt.ylabel('phi')
plt.show()
```

### 로지스틱 회귀로 이진 분류 수행하기  
---

- 불리언 인덱싱

```python
char_arr = np.array(['A', 'B', 'C', 'D', 'E'])
print(char_arr[[True, False, True, False, False]])
```

도미와 방어의 행만 골라내기

```python
bream_smelt_indexes = (train_target == 'Bream') | (train_target == 'Smelt')
train_bream_smelt = train_scaled[bream_smelt_indexes]
target_bream_smelt = train_target[bream_smelt_indexes]
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(train_bream_smelt, target_bream_smelt)
```

```python
print(lr.predict(train_bream_smelt[:5]))
```

첫 번째 열이 0에 대한 확률

```python
print(lr.predict_proba(train_bream_smelt[:5]))
```

알파벳순이라서 smelt(빙어)가 양성

```python
print(lr.classes_)
```

```python
print(lr.coef_, lr.intercept_)
```

$z=-0.404 \times(weight)-0.576\times(length)-0.663\times(diagonal)-1.013\times(height)-0.732\times(width)-2.161$ 

```python
decisions = lr.decision_function(train_bream_smelt[:5])
print(decisions)
```

```python
from scipy.special import expit

print(expit(decisions))
```

### 로지스틱 회귀로 다중 분류 수행하기  
---

LogisticRegression은 릿지 회귀와 계수의 제곱을 규제  
C는 규제를 제어하는 매개변수이고 작을수록 규제가 커짐

```python
lr = LogisticRegression(C=20, max_iter=1000)
lr.fit(train_scaled, train_target)

print(lr.score(train_scaled, train_target))
print(lr.score(test_scaled, test_target))
```

```python
print(lr.predict(test_scaled[:5]))
```

```python
proba = lr.predict_proba(test_scaled[:5])
print(np.round(proba, decimals=3))
```

```python
print(lr.classes_)
```

```python
print(lr.coef_.shape, lr.intercept_.shape)
```

5개의 특성을 사용하니까 coef 배열의 열은 5개이고 물고기 종류가 7이므로 행은 7이 된다  
z를 7개나 계산해야하는데 다중 분류는 소프트맥스라는 함수를 사용하여 각 클래스마다 z값을 계산  
가장 높은 z값을 출력하는 클래스가 예측 클래스가 된다

e-sum=$e^{z_1}+e^{z_2}+e^{z_3}+e^{z_4}+e^{z_5}+e^{z_6}+e^{z_7}$  
$\rightarrow$ s1=$\frac{e^{z_1}}{e-sum}$,s2=$\frac{e^{z_2}}{e-sum}$,s3=$\frac{e^{z_3}}{e-sum}$,s4=$\frac{e^{z_4}}{e-sum}$,s5=$\frac{e^{z_5}}{e-sum}$,s6=$\frac{e^{z_6}}{e-sum}$,s7=$\frac{e^{z_7}}{e-sum}$

```python
decision = lr.decision_function(test_scaled[:5])
print(np.round(decision, decimals=2))
```

z1~z7까지 구하고 소프트맥스 함수를 사용해 확률로 바꿈

```python
from scipy.special import softmax

proba = softmax(decision, axis=1)
print(np.round(proba, decimals=3))
```
